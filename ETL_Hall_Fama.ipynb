```python
# Bibliotecas
import pandas as pd
```

```python
# Leitura dos dataframes
df1 = pd.read_csv("./dados_brutos/data_1.csv", sep=";")
df2 = pd.read_csv("./dados_brutos/data_2.csv", sep=";")
df3 = pd.read_csv("./dados_brutos/data_3.csv", sep=";")
df4 = pd.read_csv("./dados_brutos/data_4.csv", sep=";")
```

```python
# Nome das colunas dos dataframes
print(df1.columns)
print(df2.columns)
print(df3.columns)
print(df4.columns)
```

```python
# Seleção das colunas desejadas
df1 = df1[['ANO', 'NOME', 'ESCOLA', 'CIDADE', 'ESTADO', 'PREMIO']]
df2 = df2[['ANO', 'NOME', 'ESCOLA', 'CIDADE', 'ESTADO', 'PREMIO']]

print(df1.columns)
print(df2.columns)
```

```python
# Concatenação dos dataframes
df_final = pd.concat([df1, df2, df3, df4], ignore_index=True)

#remoção de linhas duplicadas com base nas colunas 'NOME' e 'ANO', mantendo apenas a primeira ocorrência
df_final = df_final.drop_duplicates(subset=['NOME', 'ANO'], keep='first')
```

```python
# Ordenar o dataframe final pelas colunas 'ANO', 'NOME' e 'ESCOLA'
df_final.sort_values(by=['ANO', 'NOME', 'ESCOLA'], inplace=True, )

# 2. Preenche os vazios de ANO com 0
df_final['ANO'] = df_final['ANO'].fillna(0)

df_final['ANO'] = df_final['ANO'].astype(int)
df_final
```
